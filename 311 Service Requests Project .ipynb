# 4.2 Sort by avg response time

```python
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])

df['request_Closing_Time'] = df['Closed Date'] - df['Created Date']
```

```python
df = df.dropna(subset=['request_Closing_Time'])
```

```python
avg_time = df.groupby('Complaint Type')['request_Closing_Time'].mean()
```

```python
avg_time.sort_values().plot(kind='bar', figsize=(10,6))
plt.title("Avg Closing Time")
plt.show()
```

# 6. Statistical Analysis (p-value)

```python
! pip install scipy
```

```python
from scipy.stats import f_oneway
```

```python
from scipy.stats import f_oneway

groups = [group['request_Closing_Time'].values
          for name, group in df.groupby('Complaint Type')]

stat, p = f_oneway(*groups)

print("p-value:", p)
```

# 7. Kruskal-Wallis Test

```python
from scipy.stats import kruskal

stat, p = kruskal(*groups)
print("p-value:", p)
```

# 7.1Fail to reject H0: All sample distributions are equal 

```python
# Fail to reject the null hypothesis (H0). This means that all sample distributions are equal and there is no significant difference in response time across different complaint types.
```
